## all domains in sdgio ontology

```python
ld=[]
for i in range(len(list(onto.classes()))):
    a=str(list(onto.classes())[i]).split('.')[1].split('_')[0]
    if a not in ld:
        ld.append(a)
        print(str(list(onto.classes())[i]).split('.')[1].split('_')[0])
```
